```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
import pandas as pd 
df=pd.read_excel('Task1and2/train.xlsx')
```

```python
from sklearn.preprocessing import LabelEncoder  
le = LabelEncoder()
df['target'] = le.fit_transform(df['target'])
```

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
y_pred=classifier_regressor.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
df_pred=pd.read_excel('Task1and2/test.xlsx')
```

```python
result_org=classifier_regressor.predict(df_pred)
```

```python
result=pd.DataFrame({'target':result_org})
result
```

```python
file_name = 'results/Classification_logistic_regression_result.xlsx'
 
# saving the excel
result.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```
