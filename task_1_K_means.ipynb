```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df.iloc[:, :-1]
```

```python
features = df.iloc[:, :-1]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
```

```python
from kneed import KneeLocator
kneedle = KneeLocator(range(1, 11), sse, curve='convex', direction='decreasing')
optimal_k = kneedle.elbow
optimal_k
```

```python
df=df.iloc[:, :-1]
df
```

```python
optimal_k = 4  # Assume you chose k=3 from the elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
# Add the cluster labels to the original data
df['Cluster'] = clusters
```

```python
df.Cluster.describe()
```

```python
from sklearn.metrics import silhouette_score
# Calculate the silhouette score
sil_score = silhouette_score(scaled_features, clusters)
print(f'Silhouette Score: {sil_score}')
```

```python
# Reduce dimensions to 2D for visualization
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(scaled_features)

# Create a scatter plot of the clusters
plt.figure(figsize=(10, 6))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=clusters, cmap='viridis')
# Highlight the selected point
plt.scatter(reduced_features[9, 0], reduced_features[9, 1], 
            c='red', label=f'Selected Row (Cluster {clusters[9]})', edgecolors='yellow')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.title('Clusters Visualized in 2D') 
plt.show()
```

```python
highlight_cluster = 3
highlight_indices = df[df['Cluster'] == highlight_cluster].index

# Create a scatter plot of the clusters
plt.figure(figsize=(10, 6))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], c=clusters, cmap='viridis', alpha=0.6)
# Highlight the selected point
plt.scatter(reduced_features[9, 0], reduced_features[9, 1], 
            c='red', label=f'Selected Row (Cluster {clusters[9]})', edgecolors='black')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.title('Clusters Visualized in 2D')
plt.legend()
plt.show()
```
